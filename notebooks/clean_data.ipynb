```python
df = df.drop(columns=["obj_telekomTvOffer", "obj_heatingType"])
```

```python

df = df.replace('no_information', None)
```

```python
# 5. تنظيف وتحويل سرعات الإنترنت إلى Float (مع مراعاة الفواصل والنصوص مثل MBit/s)
# الكود أدناه بيحذف أي نصوص وفراغات، وبيستبدل الفاصلة بمشغل النقطة عشان البايثون يفهمها كأرقام عشريّة
for col in ['obj_telekomUploadSpeed', 'obj_telekomDownloadSpeed']:
    df[col] = df[col].astype(str).str.replace(' MBit/s', '', case=False, regex=False)
    df[col] = df[col].str.replace(',', '.', regex=False).str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce') # 'coerce' بتحول أي قيمة نصية غريبة إلى NaN عشان ما يضرب الكود
```

```python
df = df.dropna(subset=['obj_yearConstructed'])
```
